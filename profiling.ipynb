```python
import pandas as pd
import numpy as np
```

```python
data_path = '/content/drive/MyDrive/소득 예측/train.csv'
df = pd.read_csv(data_path)
```

```python
# mother 나 father 중 한명이라도 US가 있으면 Birth_Country에 (US) 추가하는 함수 정의
def add_us_if_parent_is_us(row):
    if 'US' in [row['Birth_Country (Father)'], row['Birth_Country (Mother)']]:
        return row['Birth_Country'] + '(US)'
    else:
        return row['Birth_Country']

df['Birth_Country'] = df.apply(add_us_if_parent_is_us, axis=1)

# Citizenship과 결합
df['Birth_Citizen'] = df['Birth_Country'] + '_' + df['Citizenship']

# industry가 상위범주
df['Occupation_Industry'] = df['Occupation_Status'] + "_" + df['Industry_Status']

# 근무 기간이 52주인지, 0인지, 그 사이인지 판단
def complete_work(weeks):
    if weeks == 52:
        return 'compelete'
    elif weeks == 0:
        return 'inoccupation'
    else:
        return 'incomplete'

df['Work_Type'] = df['Working_Week (Yearly)'].apply(complete_work)

# binary로 변환
def binary(value):
    if value == 0:
        return 'none'
    else:
        return 'some'

df['Dividends'] = df['Dividends'].apply(binary)
df['Gains'] = df['Gains'].apply(binary)
df['Losses'] = df['Losses'].apply(binary)

# columns 제거
# household_status는 다중공선성 때문에 제거
# losses 는 feature importance가 낮고, zero비율이 높기에 제거
df.drop(['Household_Status', 'Birth_Country (Father)', 'Citizenship',
              'Birth_Country', 'Birth_Country (Mother)'], axis=1, inplace=True)
```
